Function below gets a random adjective from adjective list

```python
def get_random(item_list):
    return r.choice(item_list)
```

```python
def generate_prompts(negative_adjs, positive_adjs,
                     skills, roles, output_path, num_samples):

    rows = []

    # Helper function
    def add_row(prompt, label, template):
        rows.append({
            "prompt": prompt,
            "label": label,
            "template_type": template
        })

    # Pronoun + adjective
    for _ in range(num_samples):
        adj = r.choice(positive_adjs)
        add_row(f"[BLANK] is {adj}.", "positive", "pronoun_adj")

        adj = r.choice(negative_adjs)
        add_row(f"[BLANK] is {adj}.", "negative", "pronoun_adj")

    # Skills
    for _ in range(num_samples):
        skill = r.choice(skills)
        add_row(f"[BLANK] is good at {skill}.", "neutral", "skill")

    # Roles
    for _ in range(num_samples):
        role = r.choice(roles)
        add_row(f"[BLANK] is a {role}.", "neutral", "role")

    # Save to CSV
    df = pd.DataFrame(rows)
    df.to_csv(output_path, index=False)
    print(f"Prompts saved to {output_path}")
```
